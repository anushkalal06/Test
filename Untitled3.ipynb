```python
import os
import glob
import numpy as np

tmp = []
target_list = []
tempDict = d.copy()

#print(tempDict)

path_to_files = os.path.join('/content/gdrive/My Drive/dataIAMFormSmall/', '*')
for filename in sorted(glob.glob(path_to_files)):
    #print(filename)
    tmp.append(filename)
    image_name = filename.split('/')[-1]
    file, ext = os.path.splitext(image_name)
    parts = file.split('-')
    form = parts[0] + '-' + parts[1]

    tempDict = d.copy()
    val = tempDict.get(form)
    if val:
        target_list.append(str(val))

img_files = np.asarray(tmp)
img_targets = np.asarray(target_list)

print(img_files.shape)
print(img_targets.shape)
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.utils import shuffle
import numpy as np
from random import sample
from PIL import Image

def generate_data(samples, target_files, batch_size=8, factor=0.1, num_classes=50):
    num_samples = len(samples)
    while True:  # Loop forever
        for offset in range(0, num_samples, batch_size):
            batch_samples = samples[offset:offset + batch_size]
            batch_targets = target_files[offset:offset + batch_size]

            images = []
            targets = []

            for i in range(len(batch_samples)):
                batch_sample = batch_samples[i]
                batch_target = batch_targets[i]

                im = Image.open(batch_sample).convert('L')  # Convert to grayscale (single channel)
                cur_width, cur_height = im.size

                height_fac = 113 / cur_height
                new_width = int(cur_width * height_fac)
                size = (new_width, 113)

                im_resized = im.resize(size, Image.LANCZOS)
                now_width, _ = im_resized.size

                # Generate random crops
                avail_x_points = list(range(0, now_width - 113))
                if len(avail_x_points) == 0:
                    start = 0  # If image width exactly matches crop width
                    im_crop = im_resized.crop((start, 0, start + 113, 113))
                    images.append(np.array(im_crop))
                    targets.append(batch_target)
                else:
                    pick_num = max(1, int(len(avail_x_points) * factor))  # Ensure at least 1 crop
                    random_startx = sample(avail_x_points, pick_num)

                    for start in random_startx:
                        im_crop = im_resized.crop((start, 0, start + 113, 113))
                        images.append(np.array(im_crop))
                        targets.append(batch_target)

            X_train = np.array(images)
            y_train = np.array(targets)

            # Print shapes (optional debugging)
            print(f"X_train shape before reshape: {X_train.shape}")
            print(f"y_train shape before one-hot encoding: {y_train.shape}")

            X_train = X_train.reshape(X_train.shape[0], 113, 113, 1)
            X_train = X_train.astype('float32') / 255.0  # Normalize

            y_train = to_categorical(y_train, num_classes=num_classes)

            print(f"X_train shape after reshape: {X_train.shape}")
            print(f"y_train shape after one-hot encoding: {y_train.shape}")

            # Shuffle X and y together
            X_train, y_train = shuffle(X_train, y_train)

            yield X_train, y_train
```

```python
train_generator = generate_data(train_files, train_targets, batch_size=batch_size, factor=0.3)
validation_generator = generate_data(validation_files, validation_targets, batch_size=batch_size, factor=0.3)
test_generator = generate_data(test_files, test_targets, batch_size=batch_size, factor=0.1)
```

```python
# A Keras model is built. Summary of the model is printed below.
def resize_image(image):
    import tensorflow as tf
    return tf.image.resize(image, [56, 56])

# Function to resize image to 64x64
row, col, ch = 113, 113, 1

model = Sequential()
model.add(ZeroPadding2D((1, 1), input_shape=(row, col, ch)))

# Resize data within the neural network
model.add(Lambda(resize_image))

# CNN model - Building the model suggested

model.add(Conv2D(filters=32, kernel_size=(5,5), strides=(2,2), padding='same', name='conv1'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), name='pool1'))

model.add(Conv2D(filters=64, kernel_size=(3,3), strides=(1,1), padding='same', name='conv2'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), name='pool2'))

model.add(Conv2D(filters=128, kernel_size=(3,3), strides=(1,1), padding='same', name='conv3'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), name='pool3'))

model.add(Flatten())
model.add(Dropout(0.5))

model.add(Dense(512, name='dense1'))
# model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(256, name='dense2'))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Assuming `num_classes` is defined
num_classes = 50  # <-- change if needed
model.add(Dense(num_classes, name='output'))
model.add(Activation('softmax'))  # softmax because output is within 50 classes
print("Worked till here")

model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])

print(model.summary())
```

```python
from keras.callbacks import ModelCheckpoint

# Training settings
epochs = 1
steps_per_epoch = 233
validation_steps = 62

# Save model checkpoints
filepath = "checkpoint2/check-{epoch:02d}-{val_loss:.4f}.keras"  # Save model after each epoch
checkpoint = ModelCheckpoint(
    filepath=filepath,
    monitor='val_loss',          # Monitor validation loss
    verbose=1,                   # Print messages when saving
    save_best_only=False,         # Save model every epoch, not just the best one
    save_weights_only=False,      # Save entire model (structure + weights)
    mode='auto'
)

callbacks_list = [checkpoint]

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    epochs=epochs,
    verbose=1,
    callbacks=callbacks_list
)
```

```python
# Load model weights
model.load_weights('checkpoint2/check-01-0.2922.keras')  # Correct path and filename

# Evaluate model
scores = model.evaluate(test_generator, steps=121)  # 121 = number of test images

# Print the accuracy like your example
print("Accuracy =", scores[1])
```
